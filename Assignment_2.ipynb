```python
import pandas as pd
```

```python
df = pd.read_csv('/content/employee_performance.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Monthly_Salary'], kde=True)  # kde=True adds a kernel density estimate
plt.title('Distribution of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Years_of_Experience', y='Performance_Score', data=df)
plt.title('Years of Experience vs. Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.show()
```

```python
numerical_features = ['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']
heatmap_data = df[numerical_features]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
